# EDA 02 — Target Variable Distribution and Division Breakdown
Analysing the distribution of Target_Days across years and divisions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from pathlib import Path

BASE = Path(".")
df24 = pd.read_csv(BASE / "STEMS_Train_2024.csv")
df25 = pd.read_csv(BASE / "STEMS_Validate_2025.csv")

print("=== TARGET_DAYS STATISTICS ===")
print(f"\n2024 Train:")
print(f"  Min  : {df24['Target_Days'].min():.2f} days")
print(f"  Max  : {df24['Target_Days'].max():.2f} days")
print(f"  Mean : {df24['Target_Days'].mean():.2f} days")
print(f"  Std  : {df24['Target_Days'].std():.2f} days")
print(f"\n2025 Validate:")
print(f"  Min  : {df25['Target_Days'].min():.2f} days")
print(f"  Max  : {df25['Target_Days'].max():.2f} days")
print(f"  Mean : {df25['Target_Days'].mean():.2f} days")
print(f"  Std  : {df25['Target_Days'].std():.2f} days")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df24['Target_Days'], bins=15, color='steelblue', edgecolor='k', alpha=0.8)
axes[0].set_title('Target_Days Distribution — 2024')
axes[0].set_xlabel('Days'); axes[0].set_ylabel('Count')
axes[0].axvline(df24['Target_Days'].mean(), color='red', linestyle='--', label=f"Mean={df24['Target_Days'].mean():.1f}")
axes[0].legend()

axes[1].hist(df25['Target_Days'], bins=15, color='darkorange', edgecolor='k', alpha=0.8)
axes[1].set_title('Target_Days Distribution — 2025')
axes[1].set_xlabel('Days'); axes[1].set_ylabel('Count')
axes[1].axvline(df25['Target_Days'].mean(), color='red', linestyle='--', label=f"Mean={df25['Target_Days'].mean():.1f}")
axes[1].legend()

plt.tight_layout()
plt.savefig('eda_02_target_distribution.png', dpi=150)
plt.show()
print("Plot saved.")
```

```python
print("=== TARGET_DAYS BY DIVISION ===")
df_all = pd.concat([df24, df25], ignore_index=True)
div_stats = df_all.groupby('Division')['Target_Days'].agg(['count','mean','std','min','max'])
print(div_stats.round(2))
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
divisions = df_all['Division'].unique()
data_by_div = [df_all[df_all['Division']==d]['Target_Days'].values for d in sorted(divisions)]
ax.boxplot(data_by_div, labels=sorted(divisions))
ax.set_title('Target_Days by Division (2024 + 2025)')
ax.set_xlabel('Division'); ax.set_ylabel('Target Days')
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('eda_02_division_boxplot.png', dpi=150)
plt.show()
```

## Observations
- AGO division has the shortest intervals (~12.8 days) — fastest harvesting fields
- LVO and UDK have the longest intervals (~15 days)
- 2025 has slightly higher variance (std=4.56) than 2024 (std=3.30)
- Distribution is roughly normal within each division
