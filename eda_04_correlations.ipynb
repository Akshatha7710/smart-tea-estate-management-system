# EDA 04 — Feature Correlations with Target_Days
Checking which features have the strongest linear relationship with our target variable.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from pathlib import Path

BASE = Path(".")
df24 = pd.read_csv(BASE / "STEMS_Train_2024.csv")
df25 = pd.read_csv(BASE / "STEMS_Validate_2025.csv")
df_all = pd.concat([df24, df25], ignore_index=True)

# Keep only numeric columns
num_df = df_all.select_dtypes(include=[np.number])
corr = num_df.corr()['Target_Days'].drop('Target_Days').sort_values()

print(" Correlation with TARGET_DAYS (sorted)")
print(corr.round(4))
```

```python
fig, ax = plt.subplots(figsize=(10, 12))
colors = ['red' if v < 0 else 'steelblue' for v in corr.values]
ax.barh(corr.index, corr.values, color=colors, edgecolor='k', alpha=0.8)
ax.axvline(0, color='black', linewidth=0.8)
ax.set_title('Feature Correlation with Target_Days')
ax.set_xlabel('Pearson Correlation Coefficient')
ax.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('eda_04_correlations.png', dpi=150)
plt.show()
```

```python
# Top positive and negative correlations
print(" Top 5 positive correlations")
print(corr.tail(5).round(4))
print("\n Top 5 negative correlations")
print(corr.head(5).round(4))
```

```python
# Full correlation heatmap for key features
key_features = ['Target_Days','Rainfall_Lag1','Rainfall_Lag2','Soil_pH',
                'Soil_Carbon','Yield_Prev_Year','Growth_Response',
                'Age_Months','Prune_Cycle_Stage','Field_Productivity']
heat = df_all[key_features].dropna().corr()
fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(heat, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(im, ax=ax)
ax.set_xticks(range(len(key_features))); ax.set_yticks(range(len(key_features)))
ax.set_xticklabels(key_features, rotation=45, ha='right')
ax.set_yticklabels(key_features)
ax.set_title('Correlation Heatmap — Key Features')
plt.tight_layout()
plt.savefig('eda_04_heatmap.png', dpi=150)
plt.show()
```

## Observations
- Most features have weak individual correlations with Target_Days (all < 0.3)
- No single feature is a strong predictor on its own
- This suggests we need a model that can combine many weak signals
- Division-level dummy variables show some of the strongest correlations
