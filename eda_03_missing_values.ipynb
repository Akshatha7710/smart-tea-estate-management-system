# EDA 03 — Missing Value Analysis
Identifying which columns have missing values and how many across all three datasets.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from pathlib import Path

BASE = Path(".")
df24 = pd.read_csv(BASE / "STEMS_Train_2024.csv")
df25 = pd.read_csv(BASE / "STEMS_Validate_2025.csv")
df26 = pd.read_csv(BASE / "STEMS_Test_2026.csv")

print("=== MISSING VALUES — 2024 TRAIN ===")
missing_24 = df24.isnull().sum()
print(missing_24[missing_24 > 0])
```

```python
print("=== MISSING VALUES — 2025 VALIDATE ===")
missing_25 = df25.isnull().sum()
print(missing_25[missing_25 > 0])
```

```python
print("=== MISSING VALUES — 2026 TEST ===")
missing_26 = df26.isnull().sum()
print(missing_26[missing_26 > 0])
```

```python
# Summary table
summary = pd.DataFrame({
    '2024_NaN': df24.isnull().sum(),
    '2025_NaN': df25.isnull().sum(),
    '2026_NaN': df26.isnull().sum(),
})
summary = summary[summary.sum(axis=1) > 0]
print("=== COMBINED MISSING VALUE SUMMARY ===")
print(summary)
```

```python
# Visualise
fig, ax = plt.subplots(figsize=(12, 6))
summary.plot(kind='bar', ax=ax, color=['steelblue','darkorange','seagreen'], edgecolor='k')
ax.set_title('Missing Values Per Column Across All Datasets')
ax.set_ylabel('Number of Missing Values')
ax.tick_params(axis='x', rotation=45)
ax.legend(['2024 Train','2025 Validate','2026 Test'])
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('eda_03_missing_values.png', dpi=150)
plt.show()
```

## Observations
- 6-7 fields missing Extent_Hect, Age_Months, Prune-related columns in both years
- Target_Lag1 and Target_Lag2 are ALL NaN in 2024 — these columns are empty and useless
- Rainfall_Current and WetDays_Current are ALL NaN in 2026 — season incomplete
- Will need imputation strategy for the 6-7 row gaps in field-level features
