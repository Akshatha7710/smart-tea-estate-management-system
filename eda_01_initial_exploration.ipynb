# EDA 01 — Initial Dataset Exploration
First look at the STEMS datasets: shapes, columns, data types and basic statistics.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from pathlib import Path

BASE = Path(".")
df24 = pd.read_csv(BASE / "STEMS_Train_2024.csv")
df25 = pd.read_csv(BASE / "STEMS_Validate_2025.csv")
df26 = pd.read_csv(BASE / "STEMS_Test_2026.csv")

print("=== DATASET SHAPES ===")
print(f"Train 2024 : {df24.shape}")
print(f"Validate 2025: {df25.shape}")
print(f"Test 2026  : {df26.shape}")
```

```python
print("=== COLUMN NAMES (44 columns) ===")
for i, col in enumerate(df24.columns, 1):
    print(f"  {i:02d}. {col}")
```

```python
print("=== DATA TYPES ===")
print(df24.dtypes)
```

```python
print("=== BASIC STATISTICS — TRAIN 2024 ===")
print(df24.describe().round(3))
```

```python
print("=== BASIC STATISTICS — VALIDATE 2025 ===")
print(df25.describe().round(3))
```

```python
print("=== FIRST 5 ROWS ===")
print(df24.head())
```

```python
print("=== DIVISION BREAKDOWN ===")
print("2024:", df24['Division'].value_counts().to_dict())
print("2025:", df25['Division'].value_counts().to_dict())
print("2026:", df26['Division'].value_counts().to_dict())
```
